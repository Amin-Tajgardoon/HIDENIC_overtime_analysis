```python
targets=['mort_icu', 'los_3']
reps=['raw', 'pca']
models=['rf', 'lr', 'nb', 'rbf-svm']
```

```python
from itertools import product
```

```python
measures=["AUC", "APR", "ECE"]
indx=pd.MultiIndex.from_tuples([(target, rep, model) for target in targets for rep in reps for model in models],
                              names=["target", "representation", "model"])
base_measures=pd.DataFrame(index=indx, columns=measures)
text_files = [f for f in os.listdir(dir_path) if f.startswith("result_hospital-overtime-style") and f.endswith(".txt")]
```

```python
for target, rep in product(targets, reps):
    for text_file in text_files:
        if(target in text_file) and ("_"+rep+"_" in text_file):
            with open(os.path.join(dir_path, text_file), "r") as f:
                all_lines=f.readlines()
            lines=[l for l in all_lines if "train_cv_score" in l]
            for model, measure in product(models, measures):
                for line in lines:
                    if (model.upper() in line) and (measure in line):
                        value=np.abs(float(line.split("[")[1].split("]")[0]))
                        base_measures.loc[(target, rep, model), measure]=value
```

```python
base_measures
```

```python
indep_tests=["CCA", "Dcorr", "RV"]
```

```python
indx=pd.MultiIndex.from_tuples([(test, target, rep, model) for test in indep_tests for target in targets for rep in reps for model in models],
                              names=["indep_test", "target", "representation", "model"])
df=pd.DataFrame(index=indx, columns=['best_prec',  'best_recall', 'best_fscore', 'best_auc', 'best_apr', 'best_thr'])
```

```python
def harmonic_mean(a,b):
    return 2*a*b/(a+b)
```

```python
from sklearn.metrics import precision_score
```

```python
for test in indep_tests:
    for target in targets:
        print(target)
        for rep in reps:
            print(rep)
            for model in models:
                best_auc=best_apr=0
                best_mean=0;best_prec=0
                for thr in np.logspace(-15,-1, num=10).tolist() + [0.05]:
                    for abs_diff_thr in np.linspace(0.001, 0.1, 10):
                        mask=bs.loc[:, idx[target, model, rep, 'AUROC', 'pval']].notna()
                        indices=bs.index[mask]
                        abs_diff=(bs.loc[indices, idx[target, model, rep, 'AUROC', 'mean']] - base_measures.loc[idx[target, rep, model], 'AUC']).abs()
                        
                        truth=((bs.loc[indices, idx[target, model, rep, 'AUROC', 'pval']] >= thr) &
                               (abs_diff >= abs_diff_thr)).values
                        fdc_pvals=fdc.loc[indices, idx[target, rep, model, test]].values
                        ## pval is the opposite of probability, i.e. lower pval means positive
                        # truth=~truth
                        prec,recall,fscore,_=[t[1] if len(t)>1 else t[0]  for t in precision_recall_fscore_support(truth, fdc_pvals>=thr)]
                        
                        try:
                            auc=roc_auc_score(truth, fdc_pvals)
                            apr=average_precision_score(truth, fdc_pvals)
                        except:
                            auc=apr=np.nan
#                         if harmonic_mean(auc,apr)>best_mean:
                        if prec>best_prec:
                            best_prec=prec
                            best_recall=recall
                            best_fscore=fscore
                            best_mean=harmonic_mean(auc,apr)
                            best_auc=auc
                            best_thr=thr
                            best_apr=apr
                            best_diff=abs_diff_thr
                print(best_diff, best_thr)
                df.loc[(test, target, rep, model), ['best_prec', 'best_recall', 'best_fscore', 'best_auc', 'best_apr', 'best_thr']] = (best_prec, best_recall, best_fscore, best_auc, best_apr, best_thr)
df=df.astype(np.float)
```

```python
indep_tests=["CCA", "Dcorr", "RV"]
targets=['mort_icu', 'los_3']
reps=['raw', 'pca']
models=['rf', 'lr', 'nb', 'rbf-svm']


indx=pd.MultiIndex.from_tuples([(test, target, rep, model) for test in indep_tests for target in targets for rep in reps for model in models],
                              names=["indep_test", "target", "representation", "model"])
df_oo=pd.DataFrame(index=indx, columns=['best_auc', 'best_apr', 'best_thr'])

for test in indep_tests:
    for target in targets:
        print(target)
        for rep in reps:
            print(rep)
            auc_list=[]; apr_list=[]
            for model in models:
                best_auc=best_apr=0
                for thr in np.logspace(-10,-1, num=10).tolist() + [0.05]:
                    mask=bs.loc[:, idx[target, model, rep, 'AUROC', 'pval']].notna()
                    indices=bs.index[mask]
                    truth=(bs.loc[indices, idx[target, model, rep, 'AUROC', 'pval']] > thr).values
                    fdc_pvals=fdc.loc[indices, idx[target, rep, model, test]].values
                    ## pval is the opposite of probability, i.e. lower pval means positive
                    # truth=~truth
                    try:
                        auc=roc_auc_score(truth, fdc_pvals)
                        apr=average_precision_score(truth, fdc_pvals)
                    except:
                        auc=apr=np.nan
                        continue
                    if auc>best_auc:
                        best_auc=auc
                        best_thr=thr
                        best_apr=apr
                        auc_list.append(auc)
                df_oo.loc[(test, target, rep, model), ['best_auc', 'best_apr', 'best_thr']] = (best_auc, best_apr, best_thr)
df_oo=df_oo.astype(np.float)
```
