## bootstrap

```python
models = ['rf', 'lr', 'rbf-svm', 'nb']
targets = ['mort_icu', 'los_3']
reps = ['raw', 'pca']
measures = ['AUROC', 'AUPRC', 'ECE']
```

```python
ind=[(target, rep, model) for target in targets for rep in reps for model in models]
ind=pd.MultiIndex.from_tuples(ind, names=['target', 'representation', 'model'])
pval_rates=pd.DataFrame(index=ind, columns=measures)
```

```python
for target in targets:
    for rep in reps:
        for model in models:
            for measure in measures:
                pvals=bs_stat.loc[:, (target,rep,model,measure,'pval')]
                rate=(pvals[pvals.notna()]<=0.01).mean()
                pval_rates.loc[(target,rep,model), measure]=rate
```

```python
pval_rates.apply(lambda x: x.astype(np.float).round(2))
```

```python
from sklearn.metrics import precision_recall_fscore_support, roc_auc_score, average_precision_score
```

```python
from sklearn.metrics import confusion_matrix
```

```python
def specificity(y_true, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    spec=tn/(fp+tn)
    return spec
```

### evaluate the class-dist-changes

```python
cdc=pd.read_pickle(os.path.join(dir_path, "class_dist_test.pkl"))
```

```python
cdc
```

```python
targets=['mort_icu', 'los_3']
reps=['raw', 'pca']
models=['rf', 'lr', 'nb', 'rbf-svm']

indx=pd.MultiIndex.from_tuples([(target, rep, model) for target in targets for rep in reps for model in models],
                              names=["target", "representation", "model"])
df=pd.DataFrame(index=indx, columns=['best_auc', 'best_apr', 'best_thr'])
```

```python
for target in targets:
    print(target)
    for rep in reps:
        print(rep)
        for model in models:
            best_auc=best_apr=0
            for thr in np.logspace(-10,-1, num=10).tolist() + [0.05]:
                mask=bs.loc[:, idx[target, model, rep, 'AUROC', 'pval']].notna()
                indices=bs.index[mask]
                truth=(bs.loc[indices, idx[target, model, rep, 'AUROC', 'pval']] > thr).values
                cdc_pvals=cdc.loc[indices, idx[target, rep, 'fisher_pval']].values
                ## pval is the opposite of probability, i.e. lower pval means positive
                # truth=~truth
                auc=roc_auc_score(truth, cdc_pvals)
                apr=average_precision_score(truth, cdc_pvals)
                if auc>best_auc:
                    best_auc=auc
                    best_thr=thr
                    best_apr=apr
                    auc_list.append(auc)
            df.loc[(target, rep, model), ['best_auc', 'best_apr', 'best_thr']] = (best_auc, best_apr, best_thr)
df=df.astype(np.float)
```

### evaluate the combination of class and feature dist. tests

```python
for target in ['mort_icu', 'los_3']:
    for rep in ['raw', 'pca']:
        for model in ['rf', 'lr', 'nb', 'rbf-svm']:
            mask=bs.loc[:, idx[target, model, rep, 'AUROC', 'pval']].notna()
            indices=bs.index[mask]
            truth=(bs.loc[indices, idx[target, model, rep, 'AUROC', 'pval']] < 0.05).values.astype(int)
            cdc_test=(cdc.loc[indices, idx[target, rep, 'fisher_pval']] < 0.05).values
            fdc_test=(fdc.loc[indices, idx[target, rep, model, 'Dcorr']] < 0.05).values
            test_vals=(cdc_test | fdc_test).astype(int)
            precision,recall,fscore,_=precision_recall_fscore_support(truth, test_vals)
            print(target, rep, model)
            print("precision={:0.2f}, recall={:0.2f}, fscore={:0.2f}".format(precision[1],recall[1],fscore[1])) 
```
